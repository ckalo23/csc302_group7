```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import joypy as jp

df=pd.read_csv("/content/drive/Shareddrives/Covid_Data/owid-covid-data (1).csv")

by_continent=df.groupby('continent')

asia=by_continent.get_group('Asia')
asia_countries=asia.groupby('location')
europe=by_continent.get_group('Europe')
europe_countries=europe.groupby('location')
africa=by_continent.get_group('Africa')
africa_countries=africa.groupby('location')
NoAm=by_continent.get_group('North America')
NoAm_countries=NoAm.groupby('location')
SoAm=by_continent.get_group('South America')
SoAm_countries=SoAm.groupby('location')
oceania=by_continent.get_group('Oceania')
oceania_countries=oceania.groupby('location')

name=["Asia","Europe","Africa","North America","South America","Oceania"]
vax=[asia['total_vaccinations'].sum(),europe['total_vaccinations'].sum(),africa['total_vaccinations'].sum(),NoAm['total_vaccinations'].sum(),SoAm['total_vaccinations'].sum(),oceania['total_vaccinations'].sum()]
deaths=[asia['total_deaths'].sum(),europe['total_deaths'].sum(),africa['total_deaths'].sum(),NoAm['total_deaths'].sum(),SoAm['total_deaths'].sum(),oceania['total_deaths'].sum()]
si=[asia_countries['stringency_index'].mean(), europe_countries['stringency_index'].mean(), africa_countries['stringency_index'].mean(), NoAm_countries['stringency_index'].mean(), SoAm_countries['stringency_index'].mean(), oceania_countries['stringency_index'].mean()]

fig, ax = plt.subplots(figsize =(20, 10))
ax.barh(name, vax)

plt.fig, axes = jp.joyplot(by_continent, column='stringency_index', by="location", ylim='own', title="Stringency Index")
plt.style.use('seaborn-white')

plt.show()

fig, axes = jp.joyplot(by_continent, column='total_cases_per_million', by="location", ylim='own', x_range=[-50000,400000], title="Total Cases Per Million People")
plt.style.use('seaborn-white')

plt.show()



```

```python
#We have a few different types of graphs on display. We can see with the bar chart how each continent fairs when it comes to 
#vaccination rates
```

```python
fig, ax = plt.subplots(figsize =(20, 10))
#ax.plot(name,vax)
ax.plot(name,deaths)
ax.set_ylabel("Deaths")
ax.set_xlabel("Continents")
#plt.show()

fig, ax = plt.subplots(figsize =(20, 10))
ax.plot(name,vax)
ax.set_ylabel("Vaccinated")
ax.set_xlabel("Continents")
```

```python
# We see that with this data, there is a certain percentage of each population on each continent thats vaccinated. We also see that
# there is a certain number of deaths on each continent. This is one way to display the data in hopes of getting an idea on how
# deaths vary across each continent and how vaccination rates also vary
```
